```python
import numpy as np 
import pandas as pd 
import nltk
import random 
import string
```

```python
import re
import matplotlib.pyplot as plt 
import matplotlib
```

```python
import nltk
nltk.download('inaugural')
from  nltk.corpus  import  inaugural
inaugural.fileids() 
inaugural.raw('1941-Roosevelt.txt') 
inaugural.raw('1961-Kennedy.txt') 
inaugural.raw('1973-Nixon.txt')
```

```python
roosevelt_speech  =  inaugural.raw('1941-Roosevelt.txt') 
kennedy_speech  =  inaugural.raw('1961-Kennedy.txt') 
nixon_speech  =  inaugural.raw('1973-Nixon.txt')
```

# 2.1 Find the number of characters, words and sentences for the mentioned documents. (Hint: use .words(), .raw(), .sent() for extracting counts):

```python
roosevelt_chars = len(roosevelt_speech) 
kennedy_chars = len(kennedy_speech) 
nixon_chars = len(nixon_speech)
print("Number  of  characters  in  Roosevelt's  speech:",  roosevelt_chars) 
print("Number  of  characters  in  Kennedy's  speech:",   kennedy_chars) 
print("Number  of  characters  in  Nixon's  speech:",  nixon_chars)
```

```python
punct = list(string.punctuation) 
punct.extend(["--", ".", ","])
```

```python
roosevelt_words=list(inaugural.words('1941-Roosevelt.txt'))
# Converting all the words to lower case
words_roosevelt = (x.lower() for x in roosevelt_words)
Cleanwords_roosevelt = [word for word in words_roosevelt if word not in punct]

roosevelt_words = (len(Cleanwords_roosevelt))

kennedy_words=list(inaugural.words('1961-Kennedy.txt')) 
words_kennedy = (x.lower() for x in kennedy_words)
Cleanwords_kennedy = [word for word in words_kennedy if word not in punct]

kennedy_words = (len(Cleanwords_kennedy))

nixon_words=list(inaugural.words('1973-Nixon.txt')) 
words_nixon = (x.lower() for x in nixon_words)
Cleanwords_nixon = [word for word in words_nixon if word not in punct]

nixon_words = (len(Cleanwords_nixon))

roosevelt_sents  =  len(inaugural.sents('1941-Roosevelt.txt')) 
kennedy_sents  =  len(inaugural.sents('1961-Kennedy.txt')) 
nixon_sents  =  len(inaugural.sents('1973-Nixon.txt'))
```

```python
print("Number  of  words  in  Roosevelt's  speech:",  roosevelt_words) 
print("Number  of  words  in  Kennedy's  speech:",  kennedy_words) 
print("Number  of  words  in  Nixon's  speech:",  nixon_words)
```

```python
print("Number  of  sentences  in  Roosevelt's  speech:",  roosevelt_sents) 
print("Number  of  sentences  in  Kennedy's  speech:",  kennedy_sents) 
print("Number  of  sentences  in  Nixon's  speech:",  nixon_sents)
```

```python
from tabulate import tabulate
# Create a list of lists with the counts for each speech
df = [
['1941-Roosevelt',  roosevelt_chars,  roosevelt_words,  roosevelt_sents], ['1961-Kennedy',  kennedy_chars,  kennedy_words,  kennedy_sents],
['1973-Nixon',  nixon_chars,  nixon_words,  nixon_sents]
]
# Create  the  table  headers
headers  =  ['Speech',  'Characters',  'Words',  'Sentences']
# Print  the  table  using  the  tabulate  library
print(tabulate(df, headers=headers))
```

```python
# Defining  a  variable  'stopwords'  which  contains  the  list of  punctuations  from‚ê£the  string  library  and  the  english  stopwords
# from  nltk
stopwords  =  nltk.corpus.stopwords.words('english')  +list(string.punctuation) 
stopwords.extend(["--", ".", ","])
```

```python
roosevelt_words=list(inaugural.words('1941-Roosevelt.txt'))
# Converting all the words to lower case
words_roosevelt = (x.lower() for x in roosevelt_words)
Filtered_roosevelt = [word for word in words_roosevelt if word not in stopwords]

Filtered_roosevelt_words  = (len(Filtered_roosevelt))

kennedy_words=list(inaugural.words('1961-Kennedy.txt')) 
words_kennedy = (x.lower() for x in kennedy_words)
Filtered_kennedy = [word for word in words_kennedy if word not in stopwords]

Filtered_kennedy_words = (len(Filtered_kennedy))

nixon_words=list(inaugural.words('1973-Nixon.txt')) 
words_nixon = (x.lower() for x in nixon_words)
Filtered_nixon = [word for word in words_nixon if word not in stopwords]

Filtered_nixon_words = (len(Filtered_nixon))
```

# 2.2 Remove all the stop words from all the three speeches.


```python
print("Word  count  after  removal  of  stopwords  in  Roosevelt's  speech:",Filtered_roosevelt_words)
print("Word  count  after  removal  of  stopwords  in  Kennedy's  speech:",Filtered_kennedy_words)
print("Word  count  after  removal  of  stopwords  in  Nixon's  speech:",Filtered_nixon_words)
```

# 2.3 Which word occurs the most number of times in his inaugural address for each president? Mention the top three words. (after removing the stop words)

```python
# Find the most common words for each president 
freq_roosevelt = nltk.FreqDist(Filtered_roosevelt) 
freq_kennedy = nltk.FreqDist(Filtered_kennedy) 
freq_nixon = nltk.FreqDist(Filtered_nixon)
```

```python
print("Top  3  common  words  in  Roosevelt's  inaugural  address:")
print(freq_roosevelt.most_common(3))

print("Top  3  common  words  in  Kennedy's  inaugural  address:") 
print(freq_kennedy.most_common(3))

print("Top  3  common  words  in  Nixon's  inaugural  address:")
print(freq_nixon.most_common(3))
```
